```python
# Create a new DataFrame by dropping rows where both specified columns have missing values
df1 = df.dropna(subset=['ORGANISATION_VERTICAL', 'ORGANISATION_SUBVERTICAL'])
# Fill all NaN values in df1 with 0
df1 = df1.fillna(0)
# Check the result
print(df1.info())  # To confirm rows have been dropped
```

```python
df1.to_csv(".csv", index=False)
```

```python
import pandas as pd

# Convert DATE_DAY to datetime format
df1['DATE_DAY'] = pd.to_datetime(df1['DATE_DAY'])

# Convert DATE_DAY to week start (Monday) and aggregate weekly
df1['Week'] = df1['DATE_DAY'].dt.to_period('W-SUN').apply(lambda r: r.start_time)

# Group by Week and aggregate numerical columns
numerical_columns = df1.select_dtypes(include=['float64', 'int64']).columns
weekly_data = df1.groupby('Week')[numerical_columns].sum().reset_index()



```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your weekly aggregated data is in the 'weekly_data' DataFrame
# and 'Week' column represents the weekly time periods

# Convert 'Week' column to datetime objects if it's not already
weekly_data['Week'] = pd.to_datetime(weekly_data['Week'])

# Filter data for weeks starting from mid-2021
start_date = pd.to_datetime('2021-09-01')  # Adjust the start date as needed
filtered_data = weekly_data[weekly_data['Week'] >= start_date]

# Columns to plot
columns_to_plot = ['GOOGLE_PAID_SEARCH_IMPRESSIONS', 'GOOGLE_SHOPPING_IMPRESSIONS',
                   'GOOGLE_PMAX_IMPRESSIONS', 'GOOGLE_DISPLAY_IMPRESSIONS',
                   'GOOGLE_VIDEO_IMPRESSIONS', 'META_FACEBOOK_IMPRESSIONS',
                   'META_INSTAGRAM_IMPRESSIONS', 'META_OTHER_IMPRESSIONS',
                   'TIKTOK_IMPRESSIONS']

# Plot the data as a stacked area chart
plt.figure(figsize=(14, 7))  # Adjust figure size as needed
plt.stackplot(filtered_data['Week'],
              [filtered_data[col] for col in columns_to_plot],
              labels=columns_to_plot)

plt.xlabel('Week')
plt.ylabel('Impressions')
plt.title('Marketing Impressions Over Week (Starting Mid-2021)')
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))  # Adjust legend position
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Columns for the pie chart
columns = ['DIRECT_CLICKS', 'BRANDED_SEARCH_CLICKS', 'ORGANIC_SEARCH_CLICKS',
           'EMAIL_CLICKS', 'REFERRAL_CLICKS', 'ALL_OTHER_CLICKS']

# Calculate the total for each column
totals = weekly_data[columns].sum()

# Create the pie chart
plt.figure(figsize=(8, 8))  # Adjust figure size as needed
plt.pie(totals, labels=totals.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Clicks by Source')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Assuming your weekly aggregated data is in the 'weekly_data' DataFrame

# Columns for the heatmap
columns = ['FIRST_PURCHASES', 'FIRST_PURCHASES_UNITS', 'FIRST_PURCHASES_ORIGINAL_PRICE',
           'FIRST_PURCHASES_GROSS_DISCOUNT', 'ALL_PURCHASES', 'ALL_PURCHASES_UNITS',
           'ALL_PURCHASES_ORIGINAL_PRICE', 'ALL_PURCHASES_GROSS_DISCOUNT']

# Calculate the correlation matrix
correlation_matrix = weekly_data[columns].corr()

# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Purchase Metrics')
plt.show()
```
